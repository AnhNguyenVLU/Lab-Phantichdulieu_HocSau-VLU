```python
#nguyenThaiAnh_187It14044
#1. Viết hàm load_data() để tải dữ liệu lên ứng dụng.Sau đó, hiển thị ra màn hình 10 dòng đầu tiên
import pandas as pd
def load_data(titanic1):
  return pd.read_csv(titanic1)
```

```python
titanic1 = load_data("titanic_disaster.csv")
```

```python
titanic1.head(10)
```

```python
#Thống kê dữ liệu thiếu trên các biến số và trực quan hóa dữ liệu thiếu bằng biểu đồ (Heat map). Hãy cho nhận xét về tình trạng thiếu dữ liệu Age, Cabin và Embarked
import seaborn as sns
sns.heatmap(titanic1.isna(),cmap='viridis')

dfage=titanic1[titanic1['Age'].isna()]
print(dfage.count())
dfcabin=titanic1[titanic1['Cabin'].isna()]
print(dfcabin.count())
dfembarked=titanic1[titanic1['Embarked'].isna()]
print(dfembarked.count())
```

```python
#3Xử lý tên cột tên Name, tách ra làm 2 cột: firstName và secondName. Lưu ý: Sau khi tách cột xong thì xóa luôn cột Name
titanic1[['Firstname', 'secondName']]=titanic1['Name'].str.split(",",expand = True)
```

```python
titanic1 = titanic1.drop('Name',axis=1)
titanic1
```

```python
#4 Xứ lý rút gọn kích thước dữ liệu trên cột Sex như sau: thay thế male  M và female  F
titanic1['Sex'].replace({'male':'M','female':'F'},inplace=True)
titanic1.head(5)
```

```python
#Xử lý dữ liệu thiếu trên biến Age bằng cách thay thế bằng giá trị trung bình tuổi: Hãy đưa
#ra quyết định dùng giá trị trung bình tuổi toàn bộ hành khách hay theo từng nhóm hạng
#vé (hạng hành khách: Pclass). Ta tiến hành làm các bước sau
#a. Sử dụng Seaborn để vẽ biểu đồ (Box plot) trực quan dữ liệu để xác định phân
#phối tuổi trên từng hạng hành khách. Nhận xét về tuổi trung bình giữa các nhóm
#hành khách. Từ đó đưa ra quyết định cách thay thế giá trị tuổi bị thiếu.
#b. Tiến hành thay thế giá trị Age bị thiếu. Sau đó, hiển thị kết quả dạng bảng và trực
#quan dữ liệu đã xử lý thiếu cho cột‘Age‘ bằng biểu đồ Heat map.

titanic1['Sex'].replace({'male':'M','female':'F'},inplace=True)
titanic1.head()
```

```python
sns.boxplot(y='Age', x='Pclass', data =titanic1)
```

```python
titanic1['Age'].fillna(round(titanic1['Age'].mean()), inplace= True)
titanic1.head(10)
```

```python
titanic1["Age"][titanic1["Pclass"] == 1] = titanic1["Age"][titanic1["Pclass"] == 1].fillna(titanic1["Age"][titanic1["Pclass"] == 1].median(), inplace = False)
titanic1["Age"][titanic1["Pclass"] == 2] = titanic1["Age"][titanic1["Pclass"] == 2].fillna(titanic1["Age"][titanic1["Pclass"] == 2].median(), inplace = False)
titanic1["Age"][titanic1["Pclass"] == 3] = titanic1["Age"][titanic1["Pclass"] == 3].fillna(titanic1["Age"][titanic1["Pclass"] == 3].median(), inplace = False)
```

```python
sns.heatmap(titanic1.isna(), 
            cmap='viridis')
```

```python
#Xây dựng biến số Agegroup có thang đo thứ tự được ánh xạ theo thang đo khoảng dựa
#trên độ tuổi của hành khách như sau: (age =< 12]  Kid; (12, 18]: Teen, (18, 60]: Adult
#và ( age > 60): Older
titanic1.loc[(titanic1['Age'] <= 12), 'Agegroup'] = 'Kid'
titanic1.loc[(titanic1['Age'] > 12) & (titanic1['Age'] <= 18), 'Agegroup'] = 'Teen'
titanic1.loc[(titanic1['Age'] > 18) & (titanic1['Age'] <= 60), 'Agegroup'] = 'Adult'
titanic1.loc[(titanic1['Age'] > 60), 'Agegroup'] = 'Older'
titanic1.head(10)
```

```python
#Tiến hành thêm đặc trưng về danh xưng (namePrefix) trong xã hội bằng cách tách Mr,
#Mrs, Miss, Master ra khỏi ―secondName‖
#secondName=titanic1['secondName']
#for i in range(len(secondName)):
#  x = str(secondName[i])
#  titanic1['namePrefix'] = secondName.apply(lambda x: x[5:] if 'Mr.' in x else (x[6:] if 'Mrs.' in x else (x[7:] if 'Miss' in x else (x[9:] if 'Master' in x else None))))
listNamePrefix=[]
for index,row in titanic1.iterrows():
    #tim khoang trang dau tien
    vitri= row['secondName'].strip().find(' ')
    secondname=row['secondName'].strip()
    #s[bd:kt]: trich 1 chuoi tu vi tri bat dau den ket thuc
    #s[bd:]: trich 1 chuoi tu vi tri bat dau den het chuoi 
    namePrefix=secondname[vitri+1:]
    listNamePrefix.append(namePrefix)

 

cotNamePrefix=pd.DataFrame(listNamePrefix,columns=['namePrefix'])
    
frameketqua=pd.concat([titanic1,cotNamePrefix],axis=1,sort=False)
   
print(frameketqua)
```

```python
#Khai thác thêm thông tin số lượng thành viên đi theo nhóm thân quen (familySize) đối với mỗi hành khách trên chuyến hải trình; family size = 1+ SibSp + Parch
titanic1['familySize'] = 1 + titanic1['SibSp'] + titanic1['Parch']
titanic1.head(10)
```

```python
 #Tạo thêm đặc trưng "Alone" để xác định hành khách đi theo nhóm hay cá nhân bằng cách dựa trên familySize như sau: Nếu familySize = 0 thì giá trị Alone = 1 và ngược lại là 0.

for i in titanic1['familySize']:
  if i == 0:
    titanic1['Alone'] = 1
  else:
    titanic1['Alone'] = 0
titanic1.head(10)
```

```python
# Tiến hành tách loại cabin (typeCabin) mà hành khách ở để lọc và phân tích đặc tính cabin. Loại cabin được kí hiệu bởi chữ cái đầu tiên. Lưu ý: Đối với dữ liệu cabin bị thiếu thì thay thế bằng "Unknown"

titanic1['typeCabin'] = titanic1['Cabin'].fillna('Unknown')
typeCabin = titanic1['typeCabin']
for i in range(len(typeCabin)):
  temp = str(typeCabin[i])
  if temp != 'Unknown':
    typeCabin[i] = temp[:1]
```

```python
#12. Trực quan thông tin tương quan tỉ lệ sống sót và thiệt mạng trên từng nhóm giới tính.
#13. Trực quan thông tin hành khách sống sót trên từng nhóm phân loại hành khách (Pclass).
#14. Trực quan thông tin hành khách sống sót trên từng nhóm giới tính và thang đo tuổi tác
#15. Trực quan xác suất hành khách sống sót dựa trên thông tin nhóm đi cùng
#16. Trực quan xác suất hành khách sống sót dựa trên thông tin giá vé
#17. Trực quan số lượng người thiệt mạng và sống sót theo phân lớp (Pclass) hành khách và
#cảng sẽ cập bến.
import matplotlib.pyplot as plt
corSurvivedSex = titanic1.groupby(['Survived','Sex'])[['Sex']].count().unstack()
print(corSurvivedSex)
corSurvivedSex['Sex'].rename(index={0:'alive',1:'dead'}).plot.pie(autopct = '%.1f %%', subplots=True, figsize=(9, 6))
```

```python
SurvivedPclass = titanic1.groupby(['Survived','Pclass'])[['Pclass']].count().unstack()
SurvivedPclass['Pclass'].rename(index={0:'alive',1:'dead'}).rename(columns={1:'1st',2:'2nd',3:'3rd'}).plot.pie(autopct = '%.1f %%', subplots=True, figsize=(12, 6))
```
