```python
import pandas as pd
from statsmodels.graphics.gofplots import qqplot
data = pd.read_csv(r'C:\Users\Thai Anh\Downloads\\babies.txt',sep='\s+')
print(data)
#taodata farme
df_cohutthuoc=data[data['smoke']==1]
print(df_cohutthuoc)

df_khonghutthuoc=data[data['smoke']==0]
print(df_khonghutthuoc)
```

```python
#trucquan
data.boxplot(by='smoke',column=['bwt'])
```

```python
#inracagiatrithongkecho2dataframe 
df_cohutthuoc.describe()
print('Kurtosic la',df_cohutthuoc['bwt'].kurtosis()+3)
print('Skew la:',df_cohutthuoc['bwt'].skew())
print('Min la:',df_cohutthuoc['bwt'].min())
print('Max la:',df_cohutthuoc['bwt'].max())
print('Mean la:',df_cohutthuoc['bwt'].mean())
print('std la:',df_cohutthuoc['bwt'].std(skipna=True))
print('Var la:',df_cohutthuoc['bwt'].var(skipna=True))
print('Q0% la:',df_cohutthuoc['bwt'].quantile(.0))
print('Q25% la:',df_cohutthuoc['bwt'].quantile(.25))
print('Q50% la:',df_cohutthuoc['bwt'].quantile(.5))
print('Q75% la:',df_cohutthuoc['bwt'].quantile(.75))
#print(IQR=Q3-Q1)
```
